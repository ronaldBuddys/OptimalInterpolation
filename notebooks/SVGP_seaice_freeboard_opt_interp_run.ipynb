```python

# ------------------------------------------------------------
# ------------------------------------------------------------
# 
#                   SCRIPT STARTS HERE
#
# ------------------------------------------------------------
# ------------------------------------------------------------


# calculate the hyper-parameters for GP on freeboard cover using a config
# - date(s)
# - window size
# - radius of inclusion
# - freeboard season

import json
import os
import numpy as np

from OptimalInterpolation import get_data_path, get_path
from OptimalInterpolation.sea_ice_freeboard import SeaIceFreeboard

```

```python



# #  change output_base_dir as needed
# output_base_dir = get_path("results")
output_base_dir = os.path.join(gdrive_mount, "MyDrive", "Dissertation")


# ---
# Parameters
# ---

# parameters that define 'tmp_dir'
grid_res = 50
incl_rad = 300
days_ahead = 4
days_behind = 4
bound_length_scales = True
prior_mean_method = "fyi_average"

min_inputs = 500
num_inducing_points = 500
min_obs_for_svgp = 4000
use_raw_data = True
initial_file_suffix = ""

# only predict on hold out - as not applying any smoothing
pred_on_hold_out = True

prediction_locations =  [
        "center_only"
        # "obs_in_cell",
        # {"name": "evenly_spaced_in_cell", "n": 100}
    ]

# --
# parameters to iterate over
# --

coarse_grids = [1]
# hold_outs = ["S3A", "S3B", "CS2_SAR", "CS2_SARIN"] 
hold_outs = [None]

# ---
# input config
# ---

config = {
    "dates": ["20190301"], #["20190201", "20190301"]
    "output_dir": os.path.join(output_base_dir, "svpg"),
    "incl_rad": incl_rad,
    "days_ahead": days_ahead,
    "days_behind": days_behind,
    "data_dir": "package",
    "season": "2018-2019",
    "grid_res": grid_res,
    "min_inputs": min_inputs,
    "verbose": 2,
    "engine": "GPflow_svgp",
    "kernel": "Matern32",
    "prior_mean_method": prior_mean_method,
    "load_params": False,
    "scale_inputs": True,
    "scale_outputs": False,
    "bound_length_scales": bound_length_scales,
    "append_to_file": True,
    "overwrite": False,
    "use_raw_data": use_raw_data,
    "store_params": True,
    "predict_locations": prediction_locations,
    "predict_in_neighbouring_cells": 1,
    "skip_if_pred_exists": True,
    "store_loss": False,
    "inducing_point_params": {
        "num_inducing_points": num_inducing_points,
        "min_obs_for_svgp": min_obs_for_svgp
    },
    "optimise_params": {
        "use_minibatch": False,
        "gamma": 1.0,
        "learning_rate": 0.07,
        "trainable_inducing_variable": False,
        "minibatch_size": 2000,
        "maxiter": 150,
        "log_freq": 10,
        "persistence": 10,
        "early_stop": True,
        "save_best": False
    }
}

# ---
# parameters
# ---

print("using config:")
print(json.dumps(config, indent=4))

# extract parameters from config

season = config.get("season", "2018-2019")
assert season == "2018-2019", "only can handle data inputs from '2018-2019' season at the moment"

gres = config.get("grid_res", 50)

# pop some config parameters
data_dir = config.pop("data_dir")
verbose = config.pop("verbose", True)
dates = config.pop("dates")

# -----
# initialise SeaIceFreeboard object
# -----

sifb = SeaIceFreeboard(grid_res=f"{gres}km",
                       length_scale_name=["x", "y", "t"],
                       verbose=verbose,
                       rng_seed=None)

# ---
# read / load data
# ---

if data_dir == "package":
    data_dir = get_data_path()

assert os.path.exists(data_dir)

sifb.load_data(aux_data_dir=os.path.join(data_dir, "aux"),
               sat_data_dir=os.path.join(data_dir, "CS2S3_CPOM"),
               raw_data_dir=os.path.join(data_dir, "RAW"),
               season=season)

# ---
# for each date increment over the model combinations
# ---

for coarse_grid_spacing in coarse_grids:
    print("|" * 100)
    print(f"coarse_grid_spacing: {coarse_grid_spacing}")
    print("-" * 10)

    for hold_out in hold_outs:
        print("|" * 100)
        print(f"hold_out: {hold_out}")
        print("-" * 10)

        for date in dates:
            print("|" * 100)
            print(f"date: {date}")
            print("-" * 10)

            # ---
            # initial run
            # ---

            print("|*" * 50)
            print(f"initial run")

            # tmp_dir is where the results will be written to
            tmp_dir = sifb.make_temp_dir(incl_rad,
                                         days_ahead,
                                         days_behind,
                                         grid_res,
                                         season,
                                         coarse_grid_spacing,
                                         hold_out,
                                         bound_length_scales,
                                         prior_mean_method)

            print(f"writing results to sub-dir:\n{tmp_dir}")
            print("-" * 10)

            # optimise - no post processing

            # by settign pred_on_hold_out=False - will create GPs for entire grid
            sifb.run(date=date,
                     optimise=True,
                     hold_out=hold_out,
                     pred_on_hold_out=pred_on_hold_out,
                     coarse_grid_spacing=coarse_grid_spacing,
                     file_suffix=initial_file_suffix,
                     post_process=None,
                     previous_results=None,
                     tmp_dir=tmp_dir,
                     **config)







```
